```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
from scipy.stats import norm
from sklearn.model_selection import train_test_split
import plotly.express as px
```

```python
df = pd.DataFrame(pd.read_csv("IrisData_full.csv"))
df.head()
```

```python
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)
df_shuffled.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Вибираємо колонки, які треба нормалізувати (числові дані)

df = df.sample(frac=1)
numeric_columns = df.select_dtypes(include=[np.number])

df[numeric_columns.columns] = (numeric_columns - numeric_columns.min()) / (numeric_columns.max() - numeric_columns.min())

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import numpy as np

train_set, test_set = train_test_split(df, test_size=0.2)

fig = px.scatter_3d(train_set, x='0', y='1',z='2', color='name', size='3')
fig.update_layout(title='Навчальна вибірка')
fig.show()
fig = px.scatter_3d(test_set, x='0', y='1',z='2', color='name', size='3')
fig.update_layout(title='Тестова вибірка')
fig.show()
# Вибрати ознаки та цільову змінну для навчального та тестового наборів
train_X = train_set[['0', '1', '2', '3']]
train_y = train_set['name']

test_X = test_set[['0', '1', '2', '3']]
test_y = test_set['name']

# Масштабування ознак
scaler = StandardScaler()
train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)

for i in range(1, 50):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(train_X, train_y)

    train_pred = knn.predict(train_X)
    test_pred = knn.predict(test_X)

    train_accuracy = accuracy_score(train_y, train_pred)
    test_accuracy = accuracy_score(test_y, test_pred)

    print(f"K = {i}; train accuracy = {train_accuracy}, test accuracy = {test_accuracy}")
```
